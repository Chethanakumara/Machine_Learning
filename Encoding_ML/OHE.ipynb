```python
import pandas as pd

# Load dataset
df = pd.read_csv("car_data_OHE.csv",index_col=[0])
print(df.head())
```

```python
df['Brand'].value_counts()
```

```python
#using pandas but pands not remember long test so we use OHE
pd.get_dummies(df,columns=['Fuel_Type','Owner_Type'])
```

```python
#n-1 
pd.get_dummies(df,columns=['Fuel_Type','Owner_Type'],drop_first=True)
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,0:8],df.iloc[:,-1] ,test_size=0.2 , random_state = 32)
```

```python
x_train
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
```

```python
from sklearn.compose import ColumnTransformer 
```

```python
col_ohe=['Brand','Fuel_Type','Transmission']
```

```python
transformer = ColumnTransformer(transformers=[
    ('tn1',SimpleImputer(),['Model_Year','Mileage_kmpl','Engine_CC','Seats']),
    ('tn2',OrdinalEncoder(categories =[['First','Second','Third']]),['Owner_Type']),
    ('tn3',OneHotEncoder(sparse_output=False,drop='first'),col_ohe)
],remainder='passthrough')
```

```python
# Apply transformation
encoded = transformer.fit_transform(x_train)
testen =transformer.transform(x_test)
```

```python
x_train.shape
```

```python
numeric_cols = ['Model_Year', 'Mileage_kmpl', 'Engine_CC', 'Seats']
ordinal_cols = ['Owner_Type']
ohe_feature_names = transformer.named_transformers_['tn3'].get_feature_names_out(col_ohe)

```

```python
all_feature_names = numeric_cols + ordinal_cols + list(ohe_feature_names)
```

```python
# Create encoded DataFrame
encoded_df = pd.DataFrame(encoded, columns=all_feature_names)
```

```python
encoded1_df = pd.DataFrame(encoded, columns=all_feature_names[:encoded.shape[1]])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
encoded_df.describe()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(encoded_df.corr(), cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap of Encoded Features")
plt.show()
```

```python
sns.pairplot(encoded_df)  # sample for performance
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

m = RandomForestRegressor(n_estimators=100, random_state=32)
m.fit(encoded, y_train)

y_pred = model.predict(testen)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("R² Score:", round(r2, 3))
print("MAE:", round(mae, 3))
print("RMSE:", round(rmse, 3))
```

```python
# 1️⃣ Actual vs Predicted
plt.figure(figsize=(7,5))
sns.scatterplot(x=y_test, y=y_pred, color='blue')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Car Prices")
plt.show()

# 2️⃣ Error Distribution
plt.figure(figsize=(7,5))
sns.histplot(y_test - y_pred, bins=20, kde=True, color='purple')
plt.title("Error Distribution (Residuals)")
plt.xlabel("Error (Actual - Predicted)")
plt.show()

# 3️⃣ Feature Importance
importances = m.feature_importances_
indices = np.argsort(importances)[::-1]
plt.figure(figsize=(10,6))
plt.title("Feature Importance in Price Prediction")
sns.barplot(x=importances[indices], y=np.array(all_feature_names)[indices])
plt.show()
```

```python
feat_imp = pd.Series(model.feature_importances_, index=encoded_df.columns)
feat_imp.nlargest(10).plot(kind='barh', color='teal')
plt.title("Top 10 Important Features")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_result = pca.fit_transform(encoded_df)

plt.figure(figsize=(8,6))
plt.scatter(pca_result[:,0], pca_result[:,1], alpha=0.6)
plt.title('PCA: Encoded Data in 2D Space')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
