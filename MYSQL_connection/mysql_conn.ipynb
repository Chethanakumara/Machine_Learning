```python
import mysql.connector
import pandas as pd 
```

```python
conn= mysql.connector.connect(
    host="localhost",
    user="root",
    password="Chetu@123",
    database="company_db"
)
```

```python
# Create cursor to execute SQL
cursor = conn.cursor()
```

```python
# Query the table
query = "SELECT experience, age, education_level, salary FROM employees"
cursor.execute(query)
```

```python
# Fetch all data
rows = cursor.fetchall()
```

```python
df = pd.DataFrame(rows,columns=['Experience', 'Age', 'Education_Level', 'Salary'])
```
