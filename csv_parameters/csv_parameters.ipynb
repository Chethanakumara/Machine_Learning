```python
#sep 
pd.read_csv('placement.csv',sep='\t')
```

```python
#names
pd.read_csv('placement.csv',names=['s_no'])
```

```python
#index_col is for if u have 2 id but one id is unnecesssary
pd.read_csv('placement.csv',index_col='sl_no')
```

```python
#header parameter
pd.read_csv('placement.csv',header=2)
```

```python
#use_cols parameter  -which column u should u take
pd.read_csv('placement.csv',usecols=['sl_no','hsc_s','workex'])
```

```python
#skiprows/nrows parameter
pd.read_csv('placement.csv',skiprows=[5,1])
```

```python
#nrow
pd.read_csv('placement.csv',nrows=10)
```

```python
#encoding
pd.read_csv('placement.csv',encoding='latin-1')
```

```python
#convertor
def rename(name):
    if name == "Commerce":
        return "com"
    else:
        return name
rename("Commerce")

```

```python
pd.read_csv('placement.csv',converters={'hsc_s':rename})
```

```python
#na_values parameters
pd.read_csv('placement.csv',na_values=['no'])
```

```python
#loding a huge dataset in chun
ddd=pd.read_csv('placement.csv',chunksize=50)
```
